First, we balance the dataset and use tfidf with SVMs.

```python
import nltk
import gensim
import csv
from nltk.corpus import stopwords
from nltk import RegexpTokenizer
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from textblob import TextBlob

tokenizer = RegexpTokenizer(r'\w+')
dlist = []
starlist = []

stop_words = set(stopwords.words('english'))

df = pd.read_csv('/home/karan/NLP/Assignment5/dataset.csv')
newdf = df.loc[:,['stars', 'text']] #copied those two columns

i = 0
onefive = pd.DataFrame(columns = ['stars', 'text'])
for index, row in newdf.iterrows():
    if row['stars'] in [1,5]:
        onefive.loc[len(onefive)] = list(row)
        
onefive.describe(include = 'all')
    
print(list(onefive['stars']).count(1))
print(list(onefive['stars']).count(5))
#set to lowercase and remove stopwords

#take 5 stars as '1' and 1 star as '0'
for index, row in onefive.iterrows():
    f = tokenizer.tokenize(row['text'].lower())
    f = [x for x in f if x not in stop_words]
    dlist.append(f)
    if(row['stars'] == 1):
        starlist.append(0)
    else:
        starlist.append(1)

from sklearn.feature_extraction.text import TfidfVectorizer


#starlist = np.asarray(starlist)


from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score
from sklearn import svm
from sklearn.svm import SVC
classifier5 = SVC()
np.asarray(starlist)

#there are 749 one-star reviews. so we can take 749 five-star reviews for a balanced dataset
main_count = 749
dlist_bal = []
starlist_bal = []
zero_count = 0
one_count = 0

for i in range(len(dlist)):
    if starlist[i] == 0 and zero_count < main_count:
        zero_count += 1
        dlist_bal.append(dlist[i])
        starlist_bal.append(starlist[i])
    elif starlist[i] == 1 and one_count < main_count:
        one_count += 1
        dlist_bal.append(dlist[i])
        starlist_bal.append(starlist[i])    
    if one_count == main_count and zero_count == main_count:
        break
        
#print(starlist_bal)

X_tfidf_bal = TfidfVectorizer(tokenizer=lambda doc: doc, lowercase=False, max_features = 15000).fit_transform(dlist_bal)
X_train, X_test, y_train, y_test = train_test_split(X_tfidf_bal, starlist_bal, test_size=0.25, random_state=101)

model1=svm.SVC(kernel='linear',C=1)
model1.fit(X_train,y_train)
scores=cross_val_score(model1,X_train,y_train,cv=10)
y_pred=model1.predict(X_test)
mn=scores.mean()
sd=scores.std()

print (" \n Decision Tree :-  ") 
print ("\n mean Accuracy: ")
print (mn)

print ("\n Standard Deviation:")
print (sd)

print ("\n Confusion Matrix : ")
# Making the Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

print ("\n Classification Report : ")
from sklearn.metrics import classification_report
print (classification_report(y_test, y_pred) )

fpr, tpr, threshold = roc_curve(y_test, y_pred)

roc_auc=roc_auc_score(y_test, y_pred)

print (roc_auc)

plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
import pandas as pd
import nltk
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.metrics import confusion_matrix
from sklearn.utils import resample
from sklearn.model_selection import cross_val_score
from sklearn.decomposition import TruncatedSVD
from textblob import TextBlob
from sklearn import svm
from sklearn import metrics
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
from nltk.stem.porter import PorterStemmer
stopwords=set(stopwords.words('english'))
import gensim
from gensim.models import KeyedVectors
from gensim.models.keyedvectors import BaseKeyedVectors
from keras.layers import Dropout
```

```python
#Importing the review dataset
dataset=pd.read_csv("dataset.csv")

#Dropping from the dataset all unnecessary columns
dataset=dataset.drop('review_id',axis=1)
dataset=dataset.drop('business_id',axis=1)
dataset=dataset.drop('date',axis=1)
dataset=dataset.drop('user_id',axis=1)
dataset=dataset.drop('type',axis=1)
dataset=dataset.drop('cool',axis=1)
dataset=dataset.drop('useful',axis=1)
dataset=dataset.drop('funny',axis=1)
uniquewords=[]
```

```python
#Preprocessing the Stars column of the dataset
for i in range(dataset.shape[0]):
        if(int(dataset.iat[i,0])<=3):
            dataset.iat[i,0]=0
        else:
            dataset.iat[i,0]=1

#Function to tokenize text and include only the meaningful tokens
def convertIntoTokens(text):
    tokenizedwords=nltk.word_tokenize(text)
    global uniquewords
    uniquewords+=tokenizedwords
    words2=nltk.pos_tag(tokenizedwords)
    words=[]
    for word,tag in words2:
        #Taking only content words - nouns,verbs, adverbs and adjectives
        if(word.isalpha() and not word in stopwords and (tag.startswith('NN') or tag.startswith('JJ') or tag.startswith('VB'))):
            words.append(word)
    return words
```

```python
#Balancing the dataset
dataset_minority=dataset[dataset.stars==0]
dataset_majority=dataset[dataset.stars==1]
dataset_minority=resample(dataset_minority,replace=True,n_samples=dataset_majority.shape[0],random_state=42)
dataset=pd.concat([dataset_minority,dataset_majority])
dataset=dataset.reset_index(drop=True)
```

```python
#Tokenizing each review
for i in range(dataset.shape[0]):
    text=str(dataset.loc[i,'text'])
    dataset.at[i,'text']=convertIntoTokens(text)
```

```python
#Appending all the reviews into a corpus
corpus=[]
for i in range(dataset.shape[0]):
    corpus.append(" ".join(list(dataset.loc[i,'text'])))
```

# PART 1

```python
#TF-IDF
vectorizer = TfidfVectorizer(min_df=2)
X=vectorizer.fit_transform(corpus)
y=dataset['stars']
```

#### CASE 1


```python
#Test-Train Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
#Support Vector Machines
model1=svm.SVC(kernel='linear',C=1)
#Training the model
model1.fit(X_train,y_train)
#Cross validation k=10
scores=cross_val_score(model1,X_train,y_train,cv=10)
#Predictions
y_predicted=model1.predict(X_test)
```

```python
#Mean accuracy
print("\nMean accuracy is",np.mean(scores))
#Standard deviation of accuracy
print("\nStandard deviation of accuracy is",np.std(scores))
cm1=confusion_matrix(y_test,y_predicted)
print("\nConfusion Matrix\n",cm1)
print("\n Classification Report")
print(classification_report(y_test,y_predicted))
fpr, tpr, threshold = roc_curve(y_test, y_predicted)
roc_auc=roc_auc_score(y_test, y_predicted)
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.3f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

### CASE 3

Using TextBlob API

```python
textblobmatrix=pd.DataFrame(columns=['subjective','objective','positive','negative','neutral','polarity','subjectivity'],index=range(dataset.shape[0]))
temp=pd.read_csv('dataset.csv')
reviews=dataset.copy()
reviews['text']=temp['text']
#reviews.drop('stars',axis=1)
#print(reviews)
for i in range(reviews.shape[0]):
    textblobmatrix.at[i,'polarity'] = 0.0
    textblobmatrix.at[i,'subjectivity'] = 0.0
    review=str(reviews.loc[i,'text'])
    sentences=nltk.sent_tokenize(review)
    count = 0
    subjective_count=0
    positive_count=0
    negative_count=0
    for sentence in sentences:
        blob = TextBlob(str(sentence))
        sentiment = blob.sentiment
        textblobmatrix.at[i,'polarity'] += sentiment[0]
        textblobmatrix.at[i,'subjectivity'] += sentiment[1]
        count +=1
        if (sentiment.polarity>0 and sentiment.subjectivity >0 ):
            positive_count +=1
        if (sentiment.polarity<0 and sentiment.subjectivity >0 ):
            negative_count +=1
        if (sentiment.subjectivity >0):
            subjective_count +=1
    obj_count=count-(subjective_count)
    textblobmatrix.at[i,'subjective']=float(subjective_count)/float(count)
    textblobmatrix.at[i,'objective']=float(obj_count)/float(count)
    textblobmatrix.at[i,'positive']=float(positive_count)/float(count)
    textblobmatrix.at[i,'negative']=float(negative_count)/float(count)
    textblobmatrix.at[i,'neutral'] = float(count-positive_count-negative_count)/float(count)
    
```

```python
y=reviews['stars']
X_train3, X_test3, y_train3, y_test3 = train_test_split(textblobmatrix, y, test_size=0.25, random_state=42)
#Support Vector Machines
model3=svm.SVC(kernel='linear',C=1)
#Training the model
model3.fit(X_train3,y_train3)
#Cross validation k=10
scores3=cross_val_score(model3,X_train3,y_train3,cv=10)
#Predictions
y_predicted3=model3.predict(X_test3)
```

```python
#Mean accuracy
print("\nMean accuracy is",np.mean(scores3))
#Standard deviation of accuracy
print("\nStandard deviation of accuracy is",np.std(scores3))
cm3=confusion_matrix(y_test3,y_predicted3)
print("\nConfusion Matrix\n",cm3)
print("\n Classification Report")
print(classification_report(y_test3,y_predicted3))
fpr, tpr, threshold = roc_curve(y_test3, y_predicted3)
roc_auc=roc_auc_score(y_test3, y_predicted3)
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.3f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

### CASE 4
Using only the words that are present in the lexicon

```python
#Converting data into the lexicon into usable format
nrcdictionary={}
with open("NRC-emotion-lexicon-wordlevel-alphabetized-v0.92.txt","r") as f:
    for line in f:
        (word,emotion,val) = line.split("\t")
        if word not in nrcdictionary.keys():
            nrcdictionary[word]={}
        nrcdictionary[word][emotion]=int(val)
```

```python
nrcf=pd.DataFrame(columns=['anger','anticipation','disgust','fear','joy','negative','positive','sadness','surprise','trust'],index=range(dataset.shape[0]))
for i in range(reviews.shape[0]):
    sentences=nltk.sent_tokenize(str(reviews.loc[i,'text']))
    nrcf[i:]=0
    words=[]
    for sentence in sentences:
        sentwords=nltk.word_tokenize(sentence)
        for j in sentwords:
            if not j in stopwords and j.isalpha():
                words.append(j)
    for word in words:
        if word in nrcdictionary.keys():
            t=nrcdictionary[word]
            for key in t.keys():
                 nrcf.at[i,key]+=int(t[key])
```

```python
#Test-Train Split
y=dataset['stars']
X_train4, X_test4, y_train4, y_test4 = train_test_split(nrcf, y, test_size=0.25, random_state=42)
#Support Vector Machines
model4=svm.SVC(kernel='linear',C=1)
#Training the model
model4.fit(X_train4,y_train4)
#Cross validation k=10
scores4=cross_val_score(model4,X_train4,y_train4,cv=10)
#Predictions
y_predicted4=model4.predict(X_test4)
```

```python
#Mean accuracy
print("\nMean accuracy is",np.mean(scores4))
#Standard deviation of accuracy
print("\nStandard deviation of accuracy is",np.std(scores4))
cm4=confusion_matrix(y_test4,y_predicted4)
print("\nConfusion Matrix\n",cm4)
print("\n Classification Report")
print(classification_report(y_test4,y_predicted4))
fpr, tpr, threshold = roc_curve(y_test4, y_predicted4)
roc_auc=roc_auc_score(y_test4, y_predicted4)
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.3f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

#### CASE 5
Using gensim to get our own word2vec

```python
model = gensim.models.Word2Vec([uniquewords], window=3, size=50, min_count=1, workers=4)
# model=model.toarray()
y=dataset['stars']
```

```python
gensimdataframe=pd.DataFrame(columns=['vector'],index=range(dataset.shape[0]))
#gensimdataframe['stars']=dataset['stars']
for i in range(dataset.shape[0]):
    gensimdataframe.at[i,'vector']=[0.0]*50
    text=dataset.loc[i,'text']
    for contentword in text:
        gensimdataframe.at[i,'vector']+=model[contentword]
```

```python
gensimdataframe=gensimdataframe['vector'].apply(pd.Series)
gensimdataframe=gensimdataframe.reset_index()
```

```python
#Test-Train Split
X_train5, X_test5, y_train5, y_test5 = train_test_split(gensimdataframe,y, test_size=0.25, random_state=42)
#Support Vector Machines
model5=svm.SVC(kernel='linear',C=1)
#Training the model
model5.fit(X_train5,y_train5)
#Cross validation k=10
scores5=cross_val_score(model5,X_train5,y_train5,cv=10)
#Predictions
y_predicted5=model5.predict(X_test5)
```

```python
#Mean accuracy
print("\nMean accuracy is",np.mean(scores5))
#Standard deviation of accuracy
print("\nStandard deviation of accuracy is",np.std(scores5))
cm5=confusion_matrix(y_test5,y_predicted5)
print("\nConfusion Matrix\n",cm5)
print("\n Classification Report")
print(classification_report(y_test5,y_predicted5))
fpr, tpr, threshold = roc_curve(y_test5, y_predicted5)
roc_auc=roc_auc_score(y_test5, y_predicted5)
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.3f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
#Importing necessary packages
import nltk
from nltk.corpus import stopwords
```

```python
#Opening the file
filename='result1.txt'
File=open(filename)


#Reading from the file
lines=File.read()
```

```python
#Tokenizing into sentences and words
sentences=nltk.sent_tokenize(lines)
words=nltk.word_tokenize(lines)
```

```python
#For stopwords
stop_words=set(stopwords.words('english'))
nouns=[]
adjectives=[]
```

```python
#Removing the punctuation marks
words2=[]
for word in words:
    if word.isalpha():
        words2.append(word)
```

```python
#After removing stopwords
words=[]
for word in words2:
    if not word in stop_words:
        words.append(word)
wordsfd=nltk.FreqDist(words)
```

```python
#Adding into the nouns and adjectives lists
taggedwords=nltk.pos_tag(words)
for word,pos in taggedwords:
    if(pos.startswith('NN')):
        nouns.append(word)
    if(pos.startswith('JJ')):
        adjectives.append(word)
```

```python
#Tokenizing each sentence into words and finding its score and storing it in a dictionary
#filteredsentences=[]
temporarysentence=[]
i=-1
sentencescores={}
for sentence in sentences:
    i+=1
    score=0
    #temporarysentence=[]
    sentencewords=nltk.word_tokenize(sentence)
    for word in sentencewords:
        if word in words:
            #temporarysentence.append(word)
            score+=wordsfd[word]
    sentencescores[i]=score
    #filteredsentences.append(temporarysentence.join(' ')
```

```python
#Printing the top 10 sentences
topsentences=[]
for key, value in sorted(sentencescores.items(), key=lambda item: (item[1], item[0]), reverse=True):
    if(len(topsentences)>=10):
        break;
    topsentences.append(sentences[key])
i=0
print("Top 10 Sentences\n ")
for sentence in topsentences:
    i+=1
    print(str(i)+". "+sentence + "\n")
```

# WordClouds

```python
import matplotlib.pyplot as plt
from wordcloud.wordcloud import WordCloud ,STOPWORDS
FrequentWords=" ".join(words)
wordcloud = WordCloud(width = 1000, height = 500,stopwords=STOPWORDS,prefer_horizontal=0.7,max_words=200,max_font_size=100).generate(FrequentWords)
plt.figure(figsize=(15,8))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
#WordCloud for Nouns
Nouns=" ".join(nouns)
wordcloud2=WordCloud(width=1000,height=500,prefer_horizontal=0.7,stopwords=STOPWORDS,max_words=200,max_font_size=100,background_color='black').generate(Nouns)
plt.figure(figsize=(15,8))
plt.imshow(wordcloud2)
plt.axis("off")
plt.show()
```

```python
#WordCloud for Adjectives
Adjectives=" ".join(adjectives)
wordcloud3=WordCloud(height=500,width=1000,prefer_horizontal=0.7,stopwords=STOPWORDS,background_color='black',max_font_size=100).generate(Adjectives)
plt.figure(figsize=(15,8))
plt.imshow(wordcloud3)
plt.axis("off")
plt.show()
```
